```python
import pandas as pd
import random

# Load existing Route and Line tables from your Excel file
file_path = r"C:\Users\86183\Downloads\Stations.xlsx"
route_df = pd.read_excel(r"C:\Users\86183\Downloads\Stations.xlsx", sheet_name='Route')  # Route table
line_df = pd.read_excel(r"C:\Users\86183\Downloads\Stations.xlsx", sheet_name='Line')  # Line table

# Merge Route and Line data to ensure proper relationships
route_line_df = route_df.merge(line_df, on='LineID', how='inner')

# Parameters
trains_per_route = 15  # Number of trains per route

# Generate Vehicle Data
vehicle_data = []
vehicle_id = 1  # Starting VehicleID

for _, route in route_line_df.iterrows():
    line_id = route['LineID']
    route_id = route['RouteID']
    for _ in range(trains_per_route):
        vehicle_data.append({
            'VehicleID': f'V{vehicle_id}',
            'LineID': line_id,
            'RouteID': route_id
        })
        vehicle_id += 1

# Convert to DataFrame
vehicle_df = pd.DataFrame(vehicle_data)

# Save to Excel or CSV
output_path = r"C:\Users\86183\Downloads\Generated_Vehicle1.xlsx"
vehicle_df.to_excel(output_path, index=False, sheet_name='Metro_Vehicles')

# Preview the DataFrame
vehicle_df.head()
```
